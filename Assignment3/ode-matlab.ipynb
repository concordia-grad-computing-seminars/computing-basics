# Integrating a System of Equations with Inputs Read From a File

To achieve this goal, you will have to complete 2 steps:
1. Read an input file (available on the repo or through moodle) to get the specific values of the parameters of integration;
2. Use an integrator (ode45, ode15s, ode23s, etc) to obtain the value of the time-series.

## System Definition

The differential system of equation:

$$ \frac{dx}{dt} = ax - bx y $$
$$ \frac{dy}{dt } = c x y - d y $$

with initial conditions:

$$ x(0) = x_0 $$
$$ y(0) = y_0 $$

is to be integrated from $t=0$ to $t = t_f$.


## Step 1: Reading input variables

The file is very simply formatted with one pair of variable:value per line, delimited by a colon.

`varname`:`value`

There are 7 values to read in for $a,b,c,d,x_0,y_0$, and $t_f$.

***Hints***:
* In matlab (newer versions 2019+), there is a function readvars() that could be easily usable to read the data in. See https://www.mathworks.com/help/matlab/ref/readvars.html.
* If you don't have access to that function, you can look at this stack overflow post for ideas on how to read from a file: https://stackoverflow.com/questions/53146773/in-octave-how-to-load-variable-and-value-from-txt-file
